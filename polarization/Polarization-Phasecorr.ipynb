```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import faraday_tools as ft
```

```python
def StokesPars(Pulse):
    I = Pulse[...,(0,3)].sum(-1) / 2
    Q = ( Pulse[...,0] - Pulse[...,3] ) / 2
    U = Pulse[...,1]
    V = Pulse[...,2]
    return I, Q, U, V
```

```python
n = np.load('data/Crab_Foldedspec.npy')
#f = np.load('/home/ramain/data/b1937/arochimeB1937+21_1024chan60ntbinfoldspec_2015-07-24T04:11:56.769+001800.0sec.npy')
#ic = np.load('/home/ramain/data/b1937/arochimeB1937+21_1024chan60ntbinicount_2015-07-24T04:11:56.769+001800.0sec.npy')
#n = f / ic[...,np.newaxis]
#n = n.sum(0)
#n -= n[:,30:40].mean(axis=1, keepdims=True)
```

```python
plt.imshow(n[:,:,(0,3)].sum(-1),aspect='auto',extent=[0,n.shape[1],400,800], cmap=cm.Greys, interpolation='nearest')
plt.xlabel('phase [rad]')
plt.ylabel('Frequency [MHz]')
```

```python
MPspec = n[:,-5:-1,:].sum(1)
#MPspec = n[:,14:21,:].sum(1)
```

```python
I, Q, U, V = StokesPars(MPspec)
freq = np.linspace(400,800,1024)
lam = (300. / freq)
P = Q + 1j*U
Z = U + 1j*V
```

```python
plt.plot(freq,I)
plt.plot(freq,Q)
plt.plot(freq,U)
plt.plot(freq,V)
plt.xlabel('Frequency [MHz]')
plt.ylabel('Stokes Pars')
```

```python
binf = 32
Ibin = I.reshape(len(I)/binf,binf).mean(-1)
Ical = np.repeat(Ibin, binf)

Qnorm = Q / Ical
Unorm = Q / Ical
Vnorm = Q / Ical

plt.plot(I)
plt.plot(Ical, 'r')
```

```python
Zphi = np.angle(Z)
Pphi = np.angle(P)

plt.subplot(211)
plt.plot(freq, np.angle(Z), 'bx')
plt.xlabel('frequency [MHz]')
plt.ylabel('Z phi')

plt.subplot(212)
plt.plot(lam**2.0, np.angle(P), 'rx')
plt.xlabel('wavelength^2')
plt.ylabel('P phi')
```

```python
vec = np.exp(-1j*Zphi)

Z1 = Z*vec            
U1 = np.real(Z1)

plt.subplot(211)
plt.plot(np.real(Z1))
plt.plot(np.imag(Z1))

plt.subplot(212)
plt.plot(lam**2.0, Q)
plt.plot(lam**2.0, U1)
```

```python
''' 
First, eliminate FR from U, to rotate between U and V properly

This will allow us to find the physical solution of zeroing V
'''

Pnorm = (Q + 1j*U1) / Ical

x=ft.FaradayTools(np.real(Pnorm))
paramsFR=x.fit_RM(-43)
print("Best fit RM:",paramsFR[1])

vecFR = np.exp(-2j * (paramsFR[1]*lam**2.0 + paramsFR[2]))
Pderot = Pnorm * vecFR
Pderot *= Ical

U_FR = np.imag(Pderot)

plt.subplot(211)
plt.plot(lam**2.0, np.real(Pnorm))
plt.plot(lam**2.0, np.real(Pderot)/Ical)
plt.plot(lam**2.0, np.std(Pnorm)*np.real(vecFR))
plt.ylabel('Stokes Q')

plt.subplot(212)
plt.plot(lam**2.0,U)
plt.plot(lam**2.0,U_FR)
plt.plot(lam**2.0, np.std(Pderot)*np.imag(vecFR))

plt.xlabel('Lambda^2')
plt.ylabel('Stokes U')
```

```python
plt.plot(lam**2.0, np.angle(Pderot), 'bx')
plt.plot(lam**2.0, np.angle(P), 'rx')

plt.ylabel('Pol angle')
plt.xlabel('lambda^2')
```

```python
"""
Update Q and U, should now have finished product
"""

Q_CD_FR = np.real(Pderot)
U_CD_FR = np.imag(Pderot)

Pfrac = abs(Pderot).sum() / I.sum()
Pangle = np.angle(Pderot.sum())

print("|P| / I = ", Pfrac)
print("Pol angle = ", Pangle)

plt.plot(freq,I)
plt.plot(freq,abs(Pderot))
```
