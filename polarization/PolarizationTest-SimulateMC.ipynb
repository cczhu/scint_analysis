```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import faraday_tools as ft
from pymc import MCMC
```

```python
def StokesPars(Pulse):
    I = Pulse[...,(0,3)].sum(-1) / 2
    Q = ( Pulse[...,0] - Pulse[...,3] ) / 2
    U = Pulse[...,1]
    V = Pulse[...,2]
    return I, Q, U, V
```

```python
def MockData(resp, q, u, v, lag, RM, phi):
    '''
    Generates mock stokes parameters, as observed at ARO
    
    resp: Frequency response of the observation
    q, u, v: Relative fraction of Q, U, V
    lag: Time delay of cables
    RM: Rotation measure
    Phi: Phase - calibrates the rotation measure
    '''
    
    IC = np.array([q,u,v])  # Q, U, V
    IC = IC / np.linalg.norm(IC)  # Normalize, so I^2 = Q^2 + U^2 + V^2

    # Define Frequency and wavelength
    freq = np.linspace(400,800,1024)
    lam = (300. / freq)
    
    # Generate flat spectra of Q, U, V, multiply by the response
    Q0 = IC[0] * resp * np.ones_like(freq)
    U0 = IC[1] * resp * np.ones_like(freq)
    V0 = IC[2] * resp * np.ones_like(freq)

    # Rotation vectors of faraday rotation and cable delay
    vecFR = np.exp(2j * (RM*lam**2.0 + phi))
    vecCD = np.exp(2j*np.pi*(np.arange(1024)/1024. * lag)) 

    # Rotate P in ISM, before reaching telescope
    P0 = Q0 + 1j*U0
    P0 *= vecFR

    # Cable delays mix measured U (U rotated between Q) and true V
    XY0 = np.imag(P0) + 1j*V0
    XY0 *= vecCD

    # Final values, as measured by the telescope
    Q = np.real(P0)
    U = np.real(XY0)
    V = np.imag(XY0)
    
    return Q, U, V
```

```python
def LSQ(data, model):
    lsq = (data - model)**2.0
    return lsq.sum()
```

```python
'''
Get Stokes I from real data, as initial condition for simulation
'''

#n = np.load('data/Crab_Foldedspec.npy')
n = np.load('data/B1937_Foldedspec.npy')

#Catalogue value of RM for initial guess
RM = 8
plt.imshow(n[:,:,(0,3)].sum(-1),aspect='auto',extent=[0,2*np.pi,400,800], interpolation='nearest',cmap=cm.inferno )
plt.xlabel('phase [rad]')
plt.ylabel('Frequency [MHz]')
```

```python
MPspec = n[:,-5:-1,:].sum(1)
#MPspec = n[:,14:21,:].sum(1)
#MPspec = n[:,60:75,:].sum(1)

I, Q, U, V = StokesPars(MPspec)
freq = np.linspace(400,800,1024)
lam = (300. / freq)
```

```python
plt.plot(freq, I, label='I')
plt.plot(freq, Q, label='Q')
plt.plot(freq, U, label='U')
plt.plot(freq, V, label='V')

plt.legend()
plt.xlabel('Frequency [MHz]')
plt.ylabel('Stokes Pars')
```

```python
'''
Sanity Check, see that I^2 = Q^2 + U^2 + V^2
'''

plt.plot(freq, I**2.0)
plt.plot(freq, Q**2.0 + U**2.0 + V**2.0)
plt.xlabel('Frequency [MHz]')
plt.ylabel('Intensity')
```

```python
# Look at the phases of P and XY
P = Q + 1j*U
XY = U + 1j*V

XYphi = np.angle(XY)
Pphi = np.angle(P)

plt.subplot(211)
plt.plot(freq, np.angle(XY), 'bx')
plt.xlabel('frequency [MHz]')
plt.ylabel('Z phi')

plt.subplot(212)
plt.plot(lam**2.0, np.angle(P), 'rx')
plt.xlabel('wavelength^2')
plt.ylabel('P phi')
```

```python
binf = 8
#Ibin = I.reshape(len(I)/binf,binf).mean(-1)  # Response of total intensity
Ibin = np.sqrt(Q**2.0 + U**2.0 + V**2.0).reshape(len(I)/binf,binf).mean(-1)  # Response of Q^2 + U^2 + V^2
Ical = np.repeat(Ibin, binf)
plt.subplot(211)
plt.plot(I)
plt.plot(Ical,'r')

plt.subplot(212)
plt.plot((Q**2.0 + U**2.0 + V**2.0) / Ical**2.0)
plt.ylim((0,2))
```

```python
"""
Fit for cable delay in V, apply to U, V

First, find a good guess for the Cable delay from the FFT
"""

XYnorm = (U + 1j*V) / Ical
FT = np.fft.rfft(np.imag(XYnorm))
FTfreq = np.fft.rfftfreq(len(V))
plt.plot(abs(FT)[:20])
plt.xlabel('Delay spacing [bins]')

lag = np.argmax(abs(FT))
print("Peak delay:", lag)
```

```python
q, u, v = MockData(Ical, 2, 1, 0, lag, RM, -0.5)

bins = 30
lag_range = np.linspace(lag-1,lag+1,bins)
RM_range = np.linspace(RM-1, RM+1, bins)
phi_range = np.linspace(-np.pi, np.pi, bins)
best = 1e10

for a in lag_range:
    for b in RM_range:
        for c in phi_range:
            q, u, v = MockData(Ical, 2, 1, 0, a, b, c)
            lsq = LSQ(Q, q) + LSQ(U, u) + LSQ(V, v)
            if lsq < best:
                best = lsq
                q_fit, u_fit, v_fit = q, u, v
                lag_fit, RM_fit, phi_fit = a, b, c
                
```

```python
print("Best fit lag, RM, phi:", lag_fit, RM_fit, phi_fit)

plt.subplot(311)
plt.plot(freq, Q)
plt.plot(freq, q_fit)
plt.ylabel('Q')

plt.subplot(312)
plt.plot(freq, U) 
plt.plot(freq, u_fit)
plt.ylabel('U')

plt.subplot(313)
plt.plot(freq, V)
plt.plot(freq, v_fit)
plt.ylabel('V')
plt.xlabel('Frequency [MHz]')
```

```python
'''
Apply first cable delay solution to XY = U + iV,
then Faraday rotation to P = Q + iU
'''

vecCD = np.exp(-2j*np.pi*(np.arange(1024)/1024. * lag_fit))
vecFR = np.exp(-2j * (RM_fit*lam**2.0 + phi_fit))
         
XYcorr = XYnorm*vecCD*Ical
U_CD = np.real(XYcorr)
V_final = np.imag(XYcorr)

Pnorm = (Q + 1j*U_CD) / Ical
Pcorr = Pnorm * vecFR * Ical

V_final = np.imag(XYcorr)
Q_final = np.real(Pcorr)
U_final = np.imag(Pcorr)

plt.plot(Q_final)
plt.plot(U_final)
plt.plot(V_final)
```
