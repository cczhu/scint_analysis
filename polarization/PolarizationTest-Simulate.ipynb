```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import faraday_tools as ft
```

```python
def StokesPars(Pulse):
    I = Pulse[...,(0,3)].sum(-1) / 2
    Q = ( Pulse[...,0] - Pulse[...,3] ) / 2
    U = Pulse[...,1]
    V = Pulse[...,2]
    return I, Q, U, V
```

```python
'''
Get Stokes I from real data, as initial condition for simulation
'''

#n = np.load('data/Crab_Foldedspec.npy')

f = np.load('/home/ramain/data/b1937/arochimeB1937+21_1024chan60ntbinfoldspec_2015-07-24T04:11:56.769+001800.0sec.npy')
ic = np.load('/home/ramain/data/b1937/arochimeB1937+21_1024chan60ntbinicount_2015-07-24T04:11:56.769+001800.0sec.npy')
n = f / ic[...,np.newaxis]
n = n.sum(0)
n -= n[:,30:40].mean(axis=1, keepdims=True)

#Catalogue value of RM for initial guess
RM = 7
plt.imshow(n[:,:,(0,3)].sum(-1),aspect='auto',extent=[0,2*np.pi,400,800], interpolation='nearest',cmap=cm.inferno )
plt.xlabel('phase [rad]')
plt.ylabel('Frequency [MHz]')
```

```python
#MPspec = n[:,-5:-1,:].sum(1)
MPspec = n[:,14:21,:].sum(1)

I, Q, U, V = StokesPars(MPspec)
freq = np.linspace(400,800,1024)
lam = (300. / freq)
```

```python
IC = np.array([3.,2.,0.])  # Q, U, V
IC /= np.linalg.norm(IC)  # Normalize, so I^2 = Q^2 + U^2 + V^2

Q0 = IC[0] * I * (np.ones_like(freq) + np.random.normal(0,1e-1,size=1024))
U0 = IC[1] * I * (np.ones_like(freq) + np.random.normal(0,1e-1,size=1024))
V0 = IC[2] * I * (np.ones_like(freq) + np.random.normal(0,1e-1,size=1024))

IC_FR = np.array([ 1, 8, -1, 0])
IC_CD = np.array([ 1, 5, 0.7, 0])

vecFR = np.exp(2j * (IC_FR[1]*lam**2.0 + IC_FR[2]))
vecCD = np.exp(2j*np.pi*(np.arange(1024)/1024. * IC_CD[1] + IC_CD[2])) 

# Rotate P in ISM, before reaching telescope
P0 = Q0 + 1j*U0
P0 *= vecFR

# Cable delays mix measured U (U rotated between Q) and true V
XY0 = np.imag(P0) + 1j*V0
XY0 *= vecCD

# Final values, as measured by the telescope
I = I
Q = np.real(P0)
U = np.real(XY0)
V = np.imag(XY0)

plt.plot(freq, I, label='I')
plt.plot(freq, Q, label='Q')
plt.plot(freq, U, label='U')
plt.plot(freq, V, label='V')

plt.legend()
plt.xlabel('Frequency [MHz]')
plt.ylabel('Stokes Pars')
```

```python
'''
Sanity Check, see that I^2 = Q^2 + U^2 + V^2
'''

plt.plot(freq, I**2.0)
plt.plot(freq, Q**2.0 + U**2.0 + V**2.0)
plt.xlabel('Frequency [MHz]')
plt.ylabel('Intensity')
```

```python
# Look at the phases of P and XY
P = Q + 1j*U
XY = U + 1j*V

XYphi = np.angle(XY)
Pphi = np.angle(P)

plt.subplot(211)
plt.plot(freq, np.angle(XY), 'bx')
plt.xlabel('frequency [MHz]')
plt.ylabel('Z phi')

plt.subplot(212)
plt.plot(lam**2.0, np.angle(P), 'rx')
plt.xlabel('wavelength^2')
plt.ylabel('P phi')
```

```python
binf = 16
Ibin = I.reshape(len(I)/binf,binf).mean(-1)
Ical = np.repeat(Ibin, binf)
plt.plot(I)
plt.plot(Ical,'r')
```

```python
"""
Fit for cable delay in V, apply to U, V

First, find a good guess for the Cable delay from the FFT
"""

XYnorm = (U + 1j*V) / Ical
FT = np.fft.rfft(np.imag(XYnorm))
FTfreq = np.fft.rfftfreq(len(V))
plt.plot(abs(FT)[:20])
plt.xlabel('Delay spacing [bins]')

lag = np.argmax(abs(FT))
print("Peak delay:", lag)
```

```python
y=ft.FaradayTools(XYnorm)
paramsCD = y.fit_RM_cabdelay(lag)
print("Best fit CD. phase:", paramsCD[1], paramsCD[2])

# Include pi/2, since we fit to V
vecCD = np.exp(-2j*np.pi*(np.arange(1024)/1024. * 
                                  paramsCD[1]) + 1j*paramsCD[2] )

RMS = 1000
lag_t = np.linspace(lag-2,lag+2,100)
lag_phi = np.linspace(-np.pi, np.pi, 100)
# Minimize RMS of V along grid of lag, phase values
for a in lag_t:
    for b in lag_phi:
        vec_t = np.exp(-2j*np.pi*(np.arange(1024)/1024. * 
                                  a) + 1j*b )
        X_t = XYnorm*vec_t
        RMS_t = np.imag(X_t).std()
        if RMS_t < RMS:
            RMS = RMS_t
            lag = a
            phi = b
            vecCD = vec_t

print("Grid Minimums of lag and phase:", lag, phi)
         
XYcorr = XYnorm*vecCD*Ical
U_CD = np.real(XYcorr)
V_final = np.imag(XYcorr)
```

```python
'''
Plot U and V before and after cable delay correction
'''

plt.subplot(211)
plt.plot(freq, np.real(XYnorm))

plt.plot(freq, np.real(XYcorr) / Ical)
plt.plot(freq, np.imag(0.5*XYnorm.std()*vecCD))
plt.ylim((-2,2))
plt.xlabel('Frequency [MHz]')
plt.ylabel('Stokes U')


plt.subplot(212)
plt.plot(freq, np.imag(XYnorm))
plt.plot(freq, np.imag(XYcorr) / Ical)
plt.plot(freq, np.real(0.5*XYnorm.std()*vecCD))
plt.ylim((-2,2))

plt.xlabel('Frequency [MHz]')
plt.ylabel('Stokes V')
```

```python
"""
Fit for RM in P, using Liam's Faraday Tools
"""

Pnorm = (Q + 1j*U_CD) / Ical

x=ft.FaradayTools(Q)
paramsFR=x.fit_RM(8)
print("Best fit RM, phase:",paramsFR[1], paramsFR[2])

vecFR = np.exp(2j * (paramsFR[1]*lam**2.0 + paramsFR[2]))
Pcorr = Pnorm * vecFR

Q_final = np.real(Pcorr)
U_final = np.real(Pcorr)
```

```python
plt.subplot(211)
plt.plot(lam**2.0, Q / Ical)
plt.plot(lam**2.0, np.real(Pcorr))
plt.plot(lam**2.0, Pcorr.std()*np.real(vecFR))
plt.ylabel('Stokes Q')
plt.ylim((-2,2))

plt.subplot(212)
plt.plot(lam**2.0, U_CD / Ical)
plt.plot(lam**2.0, np.imag(Pcorr))
plt.plot(lam**2.0, 0.5*Pcorr.std()*np.imag(vecFR))

plt.xlabel('Lambda^2')
plt.ylabel('Stokes U')
plt.ylim((-2,2))
```

```python
plt.plot(lam**2.0, np.angle(Pcorr), 'rx', label='After removing FR and CD')
plt.plot(lam**2.0, np.angle(Pnorm), 'bx', label='After removing Cable Delay')

plt.ylabel('Pol angle')
plt.xlabel('lambda^2')
```

```python
"""
Update Q and U, should now have finished product
"""

Q_CD_FR = np.real(Pcorr)
U_CD_FR = np.imag(Pcorr)

Pfrac = abs(Pcorr*Ical).sum() / I.sum()
Pangle = np.angle(Pcorr.sum())

print("|P| / I = ", Pfrac)
print("Pol angle = ", Pangle)

plt.plot(freq,I)
plt.plot(freq,abs(Pcorr*Ical))
plt.plot(freq,V_final)
```
