```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import astropy.units as u
from astropy.time import Time
import faraday_tools as ft
```

```python
def SumPulses(tlist,nfreq,npol):
    PulseAvg = np.zeros((nfreq,npol))
    for x in tlist:
        Pulse = np.load('/home/ramain/data/crab/GPlist-pulses/ARO-GPs/GP{0}.npy'.format(x.isot))
        PulseAvg += Pulse
    return PulseAvg
```

```python
def StokesPars(Pulse):
    I = Pulse[:,(0,3)].sum(-1) / 2
    Q = ( Pulse[:,0] - Pulse[:,3] ) / 2
    U = Pulse[:,1]
    V = Pulse[:,2]
    return I, Q, U, V
```

```python
def DeRotate(Q, U, V, vecCD, vecFR):
    Z = U + 1j*V
    Z *= vecCD
    U = np.real(Z)
    V = np.imag(Z)
    P = Q + 1j*U
    P *= vecFR
    Q = np.real(P)
    U = np.imag(P)
    return Q, U, V
```

```python
def RMfitting(P, RM):
    x = ft.FaradayTools(P)
    pars = x.fit_RM(RM)
    vecFR = pars[0] * np.exp(2j*(pars[1]*x.freq**2.0 + 
                                  pars[2])) + pars[3]
    return vecFR, pars
```

```python
def CDfitting(U, guess):
    x = ft.FaradayTools(U)
    pars = x.fit_RM_cabdelay(guess)
    vecCD = pars[0] * np.exp(-2j*np.pi*(np.arange(x.nfreq)/x.nfreq * 
                                  pars[1]+pars[2]))+pars[3]
    return vecCD, pars
```
