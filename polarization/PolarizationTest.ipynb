```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import faraday_tools as ft
```

```python
def StokesPars(Pulse):
    I = Pulse[...,(0,3)].sum(-1) / 2
    Q = ( Pulse[...,0] - Pulse[...,3] ) / 2
    U = Pulse[...,1]
    V = Pulse[...,2]
    return I, Q, U, V
```

```python
#n = np.load('data/Crab_Foldedspec.npy')

f = np.load('/home/ramain/data/b1937/arochimeB1937+21_1024chan60ntbinfoldspec_2015-07-24T04:11:56.769+001800.0sec.npy')
ic = np.load('/home/ramain/data/b1937/arochimeB1937+21_1024chan60ntbinicount_2015-07-24T04:11:56.769+001800.0sec.npy')
n = f / ic[...,np.newaxis]
n = n.sum(0)
n -= n[:,30:40].mean(axis=1, keepdims=True)

RM = 7
```

```python
plt.imshow(n[:,:,(0,3)].sum(-1),aspect='auto',extent=[0,2*np.pi,400,800], interpolation='nearest')
plt.xlabel('phase [rad]')
plt.ylabel('Frequency [MHz]')
```

```python
#MPspec = n[:,-5:-1,:].sum(1)
MPspec = n[:,15:21,:].sum(1)
```

```python
I, Q, U, V = StokesPars(MPspec)
freq = np.linspace(400,800,1024)
lam = (300. / freq)
P = Q + 1j*U
Z = U + 1j*V
```

```python
plt.plot(freq,I)
plt.plot(freq,Q)
plt.plot(freq,U)
plt.plot(freq,V)
plt.xlabel('Frequency [MHz]')
plt.ylabel('Stokes Pars')
```

```python
"""
Fit for cable delay in V, apply to U, V

First, find a good guess for the Cable delay from the FFT
"""

XY = U + 1j*V
FT = np.fft.rfft(V)
FTfreq = np.fft.rfftfreq(len(V))
plt.plot(abs(FT)[:200])
plt.xlabel('Delay spacing [bins]')

lag = np.argmax(abs(FT))
print("Peak delay:", lag)
```

```python
y=ft.FaradayTools(V)
paramsCD = y.fit_RM_cabdelay(lag)
print("Best fit CD:", paramsCD[1])

# Include pi/2, since we fit to V
vecCD = np.exp(-2j*np.pi*(np.arange(1024)/1024. * 
                                  paramsCD[1]+paramsCD[2]))
XYcorr = XY*vecCD

plt.subplot(211)
plt.plot(freq, np.real(XY))
plt.plot(freq, np.real(XYcorr))
plt.plot(freq, np.imag(0.5*XY.std()*vecCD))
plt.xlabel('Frequency [MHz]')
plt.ylabel('Stokes U')

plt.subplot(212)
plt.plot(freq, np.imag(XY))
plt.plot(freq, np.imag(XYcorr))
plt.plot(freq, np.real(0.5*XY.std()*vecCD))

plt.xlabel('Frequency [MHz]')
plt.ylabel('Stokes V')
```

```python
"""
Update Stokes parameters to fits from above
"""

Ucorr = np.real(XYcorr)
Vcorr = np.imag(XYcorr)
P_CD = Q + 1j*Ucorr

plt.plot(U,V, 'bx')
plt.plot(Ucorr,Vcorr, 'rx')
```

```python
"""
Fit for RM in Q, using Liam's Faraday Tools
"""

x=ft.FaradayTools(Q)
paramsFR=x.fit_RM(6.7)
print("Best fit RM:",paramsFR[1])

vecFR = np.exp(-2j * (paramsFR[1]*lam**2.0 + paramsFR[2]))
Pderot = P_CD * vecFR

plt.subplot(211)
plt.plot(lam**2.0, Q)
plt.plot(lam**2.0, np.real(Pderot))
plt.plot(lam**2.0, Pderot.std()*np.real(vecFR))
plt.ylabel('Stokes Q')

plt.subplot(212)
plt.plot(lam**2.0, U)
plt.plot(lam**2.0, np.imag(Pderot))
plt.plot(lam**2.0, Pderot.std()*np.imag(vecFR))

plt.xlabel('Lambda^2')
plt.ylabel('Stokes U')
```

```python
plt.plot(lam**2.0, np.angle(Pderot), 'bx')
plt.plot(lam**2.0, np.angle(P), 'rx')

plt.ylabel('Pol angle')
plt.xlabel('lambda^2')
```

```python
"""
Update Q and U, should now have finished product
"""

Q_CD_FR = np.real(Pderot)
U_CD_FR = np.imag(Pderot)

Pfrac = abs(Pderot).sum() / I.sum()
Pangle = np.angle(Pderot.sum())

print("|P| / I = ", Pfrac)
print("Pol angle = ", Pangle)

plt.plot(freq,I)
plt.plot(freq,abs(Pderot))
plt.plot(freq,Vcorr)
```
