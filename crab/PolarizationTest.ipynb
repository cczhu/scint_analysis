```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import faraday_tools as ft
```

```python
def StokesPars(Pulse):
    I = Pulse[...,(0,3)].sum(-1) / 2
    Q = ( Pulse[...,0] - Pulse[...,3] ) / 2
    U = Pulse[...,1]
    V = Pulse[...,2]
    return I, Q, U, V
```

```python
n = np.load('data/Crab_Foldedspec.npy')
```

```python
plt.imshow(n[:,:,(0,3)].sum(-1),aspect='auto',extent=[0,2*np.pi,400,800])
plt.xlabel('phase [rad]')
plt.ylabel('Frequency [MHz]')
```

```python
MPspec = n[:,-5:-1,:].sum(1)
#MPspec = np.load('data/MPSummed2015-07-24T08:02:22.716.npy')
```

```python
I, Q, U, V = StokesPars(MPspec)
freq = np.linspace(400,800,1024)
lam = (300. / freq)
P = Q + 1j*U
```

```python
plt.plot(freq,I)
plt.plot(freq,Q)
plt.plot(freq,U)
plt.plot(freq,V)
plt.xlabel('Frequency [MHz]')
plt.ylabel('Stokes Pars')
```

```python
"""
Fit for cable delay in V, apply to U, V

First, find a good guess for the Cable delay from the FFT
"""

XY = U + 1j*V
FT = np.fft.rfft(V)
FTfreq = np.fft.rfftfreq(len(V))
plt.plot(abs(FT)[:200])

lag = np.argmax(abs(FT))
print("Peak delay:", lag)
```

```python
y=ft.FaradayTools(V)
paramsCD = y.fit_RM_cabdelay(lag)

# Include factor of pi, since we are fitting (and setting to zero) V
vecCD = np.exp(-2j*np.pi*(np.arange(1024)/1024. * 
                                  paramsCD[1]+paramsCD[2]))
XYcorr = (XY-paramsCD[3])*vecCD + paramsCD[3]

plt.subplot(211)
plt.plot(np.real(XY))
plt.plot(np.real(XYcorr))
plt.plot(np.imag(0.05*vecCD)+paramsCD[3])

plt.subplot(212)
plt.plot(np.imag(XY))
plt.plot(np.imag(XYcorr))
plt.plot(np.real(0.05*vecCD)+paramsCD[3])

plt.xlabel('Frequency [MHz]')
plt.ylabel('Stokes U, V')
```

```python
"""
Update Stokes parameters to fits from above
"""

Ucorr = np.real(XYcorr)
Vcorr = np.imag(XYcorr)
P_CD = Q + 1j*Ucorr

plt.plot(U,V, 'bx')
plt.plot(Ucorr,Vcorr, 'rx')
```

```python
"""
Fit for RM in Q, using Liam's Faraday Tools
"""

x=ft.FaradayTools(Q)
paramsFR=x.fit_RM(-42)
print("Best fit RM:",paramsFR[1])

vecFR = np.exp(-2j * (paramsFR[1]*lam**2.0))
Pderot = P_CD * vecFR

plt.subplot(211)
plt.plot(lam**2.0, Q)
plt.plot(lam**2.0, np.real(Pderot))
plt.plot(lam**2.0, 0.05*np.real(vecFR))

plt.subplot(212)
plt.plot(lam**2.0, U)
plt.plot(lam**2.0, np.imag(Pderot))
plt.plot(lam**2.0, 0.05*np.imag(vecFR))

#plt.xlabel('Frequency [MHz]')
#plt.ylabel('Stokes Q')
```

```python
plt.plot(lam**2.0, np.angle(Pderot), 'bx')
plt.plot(lam**2.0, np.angle(P_CD), 'rx')
```

```python
"""
Update Q and U, should now have finished product
"""

Q_CD_FR = np.real(Pderot)
U_CD_FR = np.imag(Pderot)

Pfrac = abs(Pderot).sum() / I.sum()
Pangle = np.angle(Pderot.sum())

print("|P| / I = ", Pfrac)
print("Pol angle = ", Pangle)

plt.plot(freq,I)
plt.plot(freq,abs(Pderot))
plt.plot(freq,Vcorr)
```
