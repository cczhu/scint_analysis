```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
n = np.load('data/Crab_Foldedspec.npy')
```

```python
plt.imshow(n[:,:,0],aspect='auto')
```

```python
MPspec = n[:,-5:-1,:].sum(1)
#MPspec = np.load('data/MPSummed2015-07-24T08:02:22.716.npy')
```

```python
xx = MPspec[:,0]
Rexy = MPspec[:,1]
Imxy = MPspec[:,2]
yy = MPspec[:,3]
freq = np.linspace(400,800,1024)
```

```python
plt.plot(freq,Rexy)
plt.plot(freq,Imxy)
plt.xlabel('Frequency [MHz]')
plt.ylabel('Re(XY) and Im(XY)')
```

```python
plt.plot(freq, xx - yy)
plt.xlabel('Frequency [MHz]')
plt.ylabel('Stokes Q')
```

```python
#From Emily's fitting
paramsFR = np.array( [ 0.80011356,  -4.38546764e+01,   
                                    1.25454162,  -1.73598027e-02])
paramsCD = np.array([ 0.05604661, -0.26770184,
                                   -0.04650346,  0.05152873])
```

```python
freq = np.linspace(400, 800, 1024)
vecCD = paramsCD[0] * np.exp(-2j*np.pi*(np.arange(1024)/1024. * 
                                  paramsCD[1]+paramsCD[2]))+paramsCD[3]
vecFR = paramsFR[0] * np.exp(2j*(paramsFR[1]*freq**2.0 + 
                                  paramsFR[2])) + paramsFR[3]
```

```python
U = 1.*Rexy
V = 1.*Imxy
U = U * vecCD
V = V * vecCD
```

```python
"""
Plot showing removed cable delay.
Emily's fits work well here (as they should, since the
cable lengths are not changing.
"""

plt.plot(abs(U))
plt.plot(Rexy)
plt.plot(abs(V))
plt.plot(Imxy)
```

```python
Q = xx - yy
Q_derot = Q*vecFR
```

```python
"""
Removing the Faraday rotation works less well, due to changing ionosphere
and polarization angle
"""

plt.plot(Q)
plt.plot(abs(Q_derot))
```

```python
import faraday_tools as ft
```

```python
"""
Fit for RM in Q, using Liam's Faraday Tools
"""

x=ft.FaradayTools(Q)
pars=x.fit_RM(42)
vecFR = pars[0] * np.exp(2j*(pars[1]*freq**2.0 + 
                                  pars[2])) + pars[3]
plt.plot(freq, Q)
plt.plot(freq, abs(Q*vecFR))
```

```python
"""
Fit for RM and cable delay in U
"""

y=ft.FaradayTools(1.*Rexy)
pars=y.fit_RM_cabdelay(10)
vecCD = paramsCD[0] * np.exp(-2j*np.pi*(np.arange(1024)/1024. * 
                                  paramsCD[1]+paramsCD[2]))+paramsCD[3]
plt.plot(freq, 1.*Rexy)
plt.plot(freq, 1.*Rexy*U)
plt.plot(freq, 1.*Imxy)
plt.plot(freq, 1.*Imxy*U)
```
