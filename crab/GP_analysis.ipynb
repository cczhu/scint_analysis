```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from astropy.time import Time
```

```python
GPlist = np.loadtxt('giant_pulses.txt', dtype='string').T
```

```python
Times = Time(GPlist[0], format='isot', scale='utc')
SN = GPlist[1].astype(float)
phase = GPlist[2].astype(float)
```

```python
plt.ion()
```

```python
Tsort = np.argsort(Times.unix)
```

```python
ts = Times[Tsort]
t = ts.unix - ts.unix[0]
sn = SN[Tsort]
p = phase[Tsort]
```

```python
#Adjust the phases to match
p_test = 1. * p
p_test[t < 7000] += 0.22
p_test[t > 7000] += 0.31
p_test[t > 19000] -= 0.255
p_test[t > 24500] += 0.088
p_test = p_test % 1
pcorr = p_test
```

```python
plt.scatter(pcorr, t/3600, s=0.5)
plt.xlabel('phase')
plt.ylabel('time [hours]')
```

```python
plt.scatter(pcorr, np.log10(sn), s=0.5)
plt.xlabel('phase')
plt.ylabel('log(S/N)')
```

```python
#output a new giant pulse list
f = open('gp_sorted.txt','w')
for i in xrange(len(pcorr)):
    f.write('{0} {1} {2} \n'.format(ts[i], sn[i], pcorr[i]))
```

```python
#Flux distribution
plt.clf()
plt.scatter(t / 3600., np.log10(SN), marker='x', s=1, c=pcorr, cmap=cm.bwr)
plt.xlabel('time [hr]')
plt.ylabel('S/N')
```

```python
#Separate Main and Interpulse
mp_t = t[abs(pcorr-0.2) < 0.04]
mp_p = pcorr[abs(pcorr-0.2) < 0.04]
mp_sn = sn[abs(pcorr-0.2) < 0.04]
ip_t = t[abs(pcorr-0.6) < 0.04]
ip_p = pcorr[abs(pcorr-0.6) < 0.04]
ip_sn = sn[abs(pcorr-0.6) < 0.04]
```

```python
mp_x = np.histogram(np.log10(mp_sn), bins=np.linspace(0.77,3,40))[0]
ip_x = np.histogram(np.log10(ip_sn), bins=np.linspace(0.77,3,40))[0]
```

```python
plt.plot(np.linspace(0.77,3,39),np.log10(mp_x),'bx')
plt.plot(np.linspace(0.77,3,39),np.log10(ip_x),'r+')
plt.xlabel('log(S/N)')
plt.ylabel('log(Npulse)')
```

```python
y = np.log10(ip_x)[0:29]
x = np.linspace(0.77,3,39)[0:29]
from scipy.optimize import curve_fit
def line(x, m, b):
    return m*x+b

A,B = curve_fit(line,x,y)[0]
```

```python
y2 = np.log10(mp_x)[10:-1]
x2 = np.linspace(0.77,3,39)[10:-1]
from scipy.optimize import curve_fit

A2,B2 = curve_fit(line,x2,y2)[0]
```

```python
y3 = np.log10(mp_x)[0:10]
x3 = np.linspace(0.77,3,39)[0:10]
from scipy.optimize import curve_fit

A3,B3 = curve_fit(line,x3,y3)[0]
```

```python
plt.clf()
#plt.plot(x,y,'r+')
plt.plot(np.linspace(0.77,3,39),np.log10(ip_x),'r+')
plt.plot(x,line(x,A,B),'r')

plt.plot(x2,y2,'bx')
plt.plot(x2,line(x2,A2,B2))

plt.plot(x3,y3,'bx')
plt.plot(x3,line(x3,A3,B3),'b')

plt.xlabel('log(S/N)')
plt.ylabel('log(Npulse)')
```
